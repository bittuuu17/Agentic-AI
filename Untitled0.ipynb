```python
class SimpleAgent:
    def __init__(self):
        self.memory = []  # store last 3 percepts

    def perceive(self, input_data):
        self.memory.append(input_data)
        if len(self.memory) > 3:
            self.memory.pop(0)

    def decide(self):
        if "goal" in self.memory:
            return "take_action"
        return "idle"

def main():
    agent = SimpleAgent()
    percepts = ["noise", "goal", "distraction", "goal"]
    for p in percepts:
        agent.perceive(p)
        print(f"Percepts: {agent.memory} → Action: {agent.decide()}")

# Call main
if __name__ == "__main__":
    main()
```

```python
class HelloAgent:
    def __init__(self):
        self.memory = []
        self.optional = ["Hello", "World"]

    def perceive(self, input_data):
        self.memory.append(input_data)

    def reactCaseSensitive(self):
        if "hello" in self.memory:
            print("Hello")
        if "world" in self.memory:
            print("World")
        if  any(item in self.memory for item in self.optional):
            print("Hello World")

    def reactNotCaseSensitive(self):
        optional = {item.lower() for item in self.optional}
        if all(item.lower() in self.memory for item in optional):
            print("Hello World")

agent = HelloAgent()
agent.perceive("hello")
agent.perceive("world")
agent.reactCaseSensitive()
agent.reactNotCaseSensitive()




```

```python
class SimpleAgent:
    def __init__(self, goal):
        self.goal=goal
        self.state=[]
        self.done= False

    def perceive(self, input_data):
        self.state.append(input_data)

    def decide(self):
        if self.done:
            return "STOP"

        if "destination_reached" in self.state:
            self.done = True
            return "GOAL"

        if "traffic" in self.state[-1]:
            return "wait"

        if "road_clear" in self.state[-1]:
            return "move"

        return "idle"

agent1=SimpleAgent(goal="destination_reached")
environment_inputs = ["road_clear", "traffic", "road_clear", "destination_reached"]

for data in environment_inputs:
    agent1.perceive(data)
    action=agent1.decide()
    print(f"Percept: {data} → Action: {action}")
```

```python
class PlanningAgent:
    def __init__(self, goal):
        self.completed=False
        self.goal=goal
        self.plan=[]
        self.completedStepNumber=0

    def create_plan(self):
        if self.goal == "Go To Office":
            self.plan=[
                "Get Ready",
                "Leave Home",
                "Take Transport",
                "Get To Work",
                "Go To Office"
            ]

    def perceive(self):
        if self.completed==True:
            return "Done"
        return self.plan[self.completedStepNumber]

    def act(self):
        action = self.plan[self.completedStepNumber]
        print(f"Executing Now : {action}")
        self.completedStepNumber+=1

        if self.completedStepNumber >= len(self.plan):
            self.completed=True
            print("Goal Achieved")

def main():
    agent= PlanningAgent(goal="Go To Office")
    agent.create_plan()

    while not agent.completed:
        percept=agent.perceive()
        agent.act()

if __name__ == "__main__":
    main()
```

# Day 5 – LLM-Powered Planning Agent (Perplexity)

## Overview
Day 5 marks the transition from rule-based agents to a **true agentic AI system**.  
In this stage, the agent no longer relies on hardcoded logic to decide actions. Instead, it uses a **Large Language Model (Perplexity – sonar-pro)** as a planner to dynamically select tools based on the current goal and internal state.

This architecture follows the principle:

> **LLMs reason, tools execute, state remembers**

---

## Objective
Build an agent that:
- Uses an LLM to decide which tool to invoke
- Maintains internal memory (state)
- Executes deterministic tools
- Separates reasoning from execution
- Aligns with modern agent frameworks and MCP concepts

---

## Agent Architecture




```python
import requests
from google.colab import userdata


API_KEY = userdata.get("PERPLEXITY_API_KEY")

def perplexity_decided_tool(goal, state):
    # Perplexity requires "Bearer <token>" format
    headers = {
        "Authorization": f"Bearer {API_KEY}",
        "Content-Type": "application/json",
        "accept": "application/json"
    }

    # Perplexity uses the Chat Completions format
    messages = [
        {"role": "system", "content": "You are a specialized tool-decision agent. Return only the tool name."},
        {"role": "user", "content": f"Goal: {goal}\nCurrent state: {state}\nDecide tool: get_weather / send_message / none."}
    ]

    data = {
        "model": "sonar-pro", # Valid 2025 model
        "messages": messages,
        "temperature": 0  # Set to 0 for consistent tool selection
    }

    # Use the correct API endpoint
    response = requests.post(
        "https://api.perplexity.ai/chat/completions",
        headers=headers,
        json=data
    )

    if response.status_code == 200:
        result = response.json()
        # Extract response text from the chat completion object
        return result["choices"][0]["message"]["content"].strip().lower()
    else:
        print(f"Error {response.status_code}: {response.text}")
        return "none"



def get_weather(city):
  weather_date={"New York" : "rainy", "Mumbai" : "sunny"}
  return weather_date.get(city, "not determined")

def send_message(msg):
    print(f"Sending Message: {msg}")

class LLMPlanningAgent:

    def __init__(self, city, goal):
        self.city = city
        self.goal = goal
        self.state = {}

    def decide_tool(self):
        print("deciding tool")
        return perplexity_decided_tool(self.goal, self.state)

    def act(self):
        print(f"Planning to {self.goal}")
        tool = self.decide_tool()
        print(f"Using tool choosen by perplexity: {tool}")

        if tool == "get_weather":
            self.state["weather"]= get_weather(self.city)
            print(f"Weather in {self.city} is {self.state['weather']}")
        elif tool == "send_message":
            if "weather" in self.state:
                 message = f"Weather in {self.city} is {self.state['weather']}"
                 send_message(message)
        else:
            print("no action")


def main():
    agent = LLMPlanningAgent(goal="Check weather", city="New York")
    agent.act()

if __name__ == "__main__":
    main()
```

```python
import requests
from google.colab import userdata

API_KEY = userdata.get("PERPLEXITY_API_KEY")

def perplexity_decided_tool(goal, state):

  headers= {
      "Authorization": f"Bearer {API_KEY}",
      "Content-Type": "application/json",
      "accept": "application/json"
  }

  messages=[
      {
          "role": "system",
          "content":(
              " you are a planning agent"
              "Return only the tool name. based on goal"
              "get_weather, send_message, done"
          )
      },
      {
          "role": "user",
          "content": f"Goal: {goal}\nState: {state}"
      }
  ]

  data= {
      "model" : "sonar-pro",
      "messages": messages,
      "temperature": 0
  }

  response = requests.post(
      "https://api.perplexity.ai/chat/completions",
      headers=headers,
      json=data
  )

  return response.json()["choices"][0]["message"]["content"].strip().lower()


def get_weather(city):
    weather_data={"New York": "rainy", "Mumbai": "sunny"}
    return weather_data.get(city, "not determined")

def send_message(msg):
    print(f"Sending Message: {msg}")

class Agent:
    def __init__(self, city, goal):
        self.city=city
        self.goal=goal
        self.state={
            "weather": None,
            "message_sent": False,
            "steps":[]
        }

    def act(self):
          print("Agent Decision Loop Starts")

          while True:
              tool= perplexity_decided_tool(self.goal,self.state)
              print(f"Choosen Toon: {tool}")
              self.state["steps"].append(tool)

              if tool== "get_weather":
                  self.state["weather"]= get_weather(self.city)
                  print(f"Weather in {self.city} is {self.state['weather']}")
              elif tool == "send_message":
                  if "weather" in self.state and not self.state["message_sent"]:
                      send_message(f"Weather in {self.city} is {self.state['weather']}")
                      self.state["message_sent"] = True

              elif tool =="done":
                  print("Goal Achieved")
                  break
              else:
                  print("No action")
                  break

def main():
    agent = Agent(goal="advice user based on weather", city="New York")
    agent.act()


if __name__ == "__main__":
    main()
```
